```python
import pandas as pd
import numpy as np
import time
from datetime import date
from sklearn.ensemble import RandomForestClassifier
from sklearn import preprocessing
t_user  =pd.read_csv('D:\研究生\Chaos\JD\信贷需求/t_user.csv')
t_loan = pd.read_csv('D:\研究生\Chaos\JD\信贷需求/t_loan.csv')
```

```python

def original_data(x):
    result = 5**x-1
    return result
def get_month(x):
    timeArray = time.strptime(x,"%Y-%m-%d %H:%M:%S")
    return timeArray.tm_mon

        
t_loan['origin_loan_amount'] = t_loan.loan_amount.apply(original_data)
t_loan['month'] = t_loan.loan_time.apply(get_month)
t = t_loan
```

```python
#计算8,9,10,11月份，12月份当月要还多少钱
#计算8,9,10,11月份，12月份时一共要还多少钱


def month_9_topay(df):
    loan_month = df['month']
    plannum = df['plannum']
    loan_amount = df['origin_loan_amount']
    if loan_month+plannum>=9:
        loan_to_pay = loan_amount/plannum
    else:
        loan_to_pay = 0
    return loan_to_pay

def month_9_topay_future(df):
    loan_month = df['month']
    plannum = df['plannum']
    loan_amount = df['origin_loan_amount']
    if loan_month+plannum>=9:
        loan_to_pay_future = loan_amount-(loan_amount/plannum)*(9-loan_month-1)
    else:
        loan_to_pay_future = 0
    return loan_to_pay_future
t_loan = t[t.month<9]
t_loan['month_9_topay'] = t_loan.apply(month_9_topay,axis=1)
t_loan['month_9_topay_future'] = t_loan.apply(month_9_topay_future,axis=1)

t_loan_topay9 = t_loan[['uid','month_9_topay','month_9_topay_future']]
t_loan_topay9 = t_loan_topay9.groupby('uid').sum().reset_index()
t_loan_topay9.columns=['uid','topay_thismonth9','topay_future9']


def month_10_topay(df):
    loan_month = df['month']
    plannum = df['plannum']
    loan_amount = df['origin_loan_amount']
    if loan_month+plannum>=10:
        loan_to_pay = loan_amount/plannum
    else:
        loan_to_pay = 0
    return loan_to_pay

def month_10_topay_future(df):
    loan_month = df['month']
    plannum = df['plannum']
    loan_amount = df['origin_loan_amount']
    if loan_month+plannum>=10:
        loan_to_pay_future = loan_amount-(loan_amount/plannum)*(10-loan_month-1)
    else:
        loan_to_pay_future = 0
    return loan_to_pay_future
t_loan = t[t.month<10]
t_loan['month_10_topay'] = t_loan.apply(month_10_topay,axis=1)
t_loan['month_10_topay_future'] = t_loan.apply(month_10_topay_future,axis=1)

t_loan_topay10 = t_loan[['uid','month_10_topay','month_10_topay_future']]
t_loan_topay10 = t_loan_topay10.groupby('uid').sum().reset_index()
t_loan_topay10.columns=['uid','topay_thismonth10','topay_future10']




def month_11_topay(df):
    loan_month = df['month']
    plannum = df['plannum']
    loan_amount = df['origin_loan_amount']
    if loan_month+plannum>=11:
        loan_to_pay = loan_amount/plannum
    else:
        loan_to_pay = 0
    return loan_to_pay

def month_11_topay_future(df):
    loan_month = df['month']
    plannum = df['plannum']
    loan_amount = df['origin_loan_amount']
    if loan_month+plannum>=11:
        loan_to_pay_future = loan_amount-(loan_amount/plannum)*(11-loan_month-1)
    else:
        loan_to_pay_future = 0
    return loan_to_pay_future
t_loan = t[t.month<11]
t_loan['month_11_topay'] = t_loan.apply(month_11_topay,axis=1)
t_loan['month_11_topay_future'] = t_loan.apply(month_11_topay_future,axis=1)

t_loan_topay11 = t_loan[['uid','month_11_topay','month_11_topay_future']]
t_loan_topay11 = t_loan_topay11.groupby('uid').sum().reset_index()
t_loan_topay11.columns=['uid','topay_thismonth','topay_future']
result = t_user[['uid']]
result = pd.merge(result,t_loan_topay11,on=['uid'],how='left')
result = pd.merge(result,t_loan_topay9,on=['uid'],how='left')
result = pd.merge(result,t_loan_topay10,on=['uid'],how='left')
result.rename(columns={'topay_thismonth10':'topay_thismonth1', 'topay_thismonth9':'topay_thismonth2', 'topay_future10': 'topay_future1', 'topay_future9': 'topay_future2'}, inplace=True) 
result = result.fillna(0)
result.to_csv('D:/研究生/Chaos/JD/信贷需求/feature11/t_loan_topay11.csv',index=None) 



def month_12_topay(df):
    loan_month = df['month']
    plannum = df['plannum']
    loan_amount = df['origin_loan_amount']
    if loan_month+plannum>=12:
        loan_to_pay = loan_amount/plannum
    else:
        loan_to_pay = 0
    return loan_to_pay
def month_12_topay_future(df):
    loan_month = df['month']
    plannum = df['plannum']
    loan_amount = df['origin_loan_amount']
    if loan_month+plannum>=12:
        loan_to_pay_future = loan_amount-(loan_amount/plannum)*(12-loan_month-1)
    else:
        loan_to_pay_future= 0
    return loan_to_pay_future
t_loan['month_12_topay'] = t_loan.apply(month_12_topay,axis=1)
t_loan['month_12_topay_future'] = t_loan.apply(month_12_topay_future,axis=1)
t_loan_topay12 = t_loan[['uid','month_12_topay','month_12_topay_future']]
t_loan_topay12 = t_loan_topay12.groupby('uid').sum().reset_index()
t_loan_topay12.columns=['uid','topay_thismonth','topay_future']
              
result = t_user[['uid']]           
result = pd.merge(result,t_loan_topay11,on=['uid'],how='left')
result.rename(columns={'topay_thismonth': 'topay_thismonth1','topay_future': 'topay_future1'}, inplace=True) 
result = pd.merge(result,t_loan_topay12,on=['uid'],how='left')           
result = pd.merge(result,t_loan_topay10,on=['uid'],how='left')
result.rename(columns={'topay_thismonth10': 'topay_thismonth2', 'topay_future10': 'topay_future2',}, inplace=True) 
result = result.fillna(0)
result = result[['uid','topay_thismonth','topay_future','topay_thismonth2','topay_future2','topay_thismonth1','topay_future1']]
result.to_csv('D:/研究生/Chaos/JD/信贷需求/feature12/t_loan_topay12.csv',index=None) 
```

```python
#计算每个月的limit
loan_by_month  = pd.read_csv('D:/研究生/Chaos/JD/信贷需求/feature/loan_by_month_sum.csv')
t_user  = pd.read_csv('D:/研究生/Chaos/JD/信贷需求/feature/user_active.csv')
limit = t_user[['uid','limit']]
limit = pd.merge(limit,loan_by_month,on=['uid'],how='left')
limit = pd.merge(limit,t_loan_topay9,on=['uid'],how='left')
limit = pd.merge(limit,t_loan_topay10,on=['uid'],how='left')
limit = pd.merge(limit,t_loan_topay11,on=['uid'],how='left')
def limit8(x):
    limit_start = x['limit']
    future = x['loan_8_sum']
    if future > limit_start:
        result = future
    else:
        result = limit_start +500
    return result

def limit9(x):
    limit_8 = x['limit_8']
    future = x['topay_future9']
    topay = x['topay_thismonth9']
    if (future-topay) > limit_8:
        result = future - topay
    else:
        result = limit_8 +500
    return result

def limit10(x):
    limit_9 = x['limit_9']
    future = x['topay_future10']
    topay = x['topay_thismonth10']
    if (future-topay) > limit_9:
        result = future - topay
    else:
        result = limit_9 +500
    return result

def limit11(x):
    limit_10 = x['limit_10']
    future = x['topay_future']
    topay = x['topay_thismonth']
    if (future-topay) > limit_10:
        result = future - topay
    else:
        result = limit_10 +500
    return result



limit['limit_8'] = limit.apply(limit8, axis=1)
limit['limit_9'] = limit.apply(limit9, axis=1)
limit['limit_10'] = limit.apply(limit10, axis=1)
limit['limit_11'] = limit.apply(limit11, axis=1)



new_limit_11 = limit[['uid','limit_8','limit_9','limit_10']]
new_limit_11.columns = ['uid','limit_3','limit_2','limit_1']
new_limit_11.to_csv('D:/研究生/Chaos/JD/信贷需求/feature11/new_limit_11.csv',index=None) 




new_limit_12 = limit[['uid','limit_9','limit_10','limit_11']]
new_limit_12.columns = ['uid','limit_3','limit_2','limit_1']
new_limit_12.to_csv('D:/研究生/Chaos/JD/信贷需求/feature12/new_limit_12.csv',index=None) 
```
