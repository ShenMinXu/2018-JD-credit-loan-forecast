```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
def original_data(x):
    result = 5**x-1
    return result
# feature1: sum 11
'''t_loan'''
raw_loan_data = pd.read_csv('D:\研究生\Chaos\JD\信贷需求/t_loan.csv')
raw_loan_data['loan_time'] = raw_loan_data['loan_time'].apply(lambda x: x.split('-')[1])
raw_loan_data['loan_amount'] = raw_loan_data['loan_amount'].apply(original_data)
raw_loan_data = raw_loan_data[(raw_loan_data.loan_time != '12')&(raw_loan_data.loan_time != '11')]
new_loan_data = raw_loan_data.groupby('uid', sort=False)\
	.agg({'loan_time': len, 'loan_amount':lambda x: sum(x), 'plannum': lambda y: sum(y)})\
	.rename(columns={'loan_time': 'loan_count_sum', 'loan_amount': 'loan_price_sum', 'plannum': 'plannum_sum'})
new_loan_data.to_csv('./featureData/new_loan_data.csv')
#
user_data = pd.read_csv('D:\研究生\Chaos\JD\信贷需求/t_user.csv')
loan_data = pd.read_csv('./featureData/new_loan_data.csv')
merge_loan_data = pd.merge(user_data, loan_data, how='left', on='uid')
merge_loan_data.to_csv('./featureData/merge_loan_data.csv')
'''t_click'''
raw_loan_data = pd.read_csv('D:\研究生\Chaos\JD\信贷需求/t_click.csv')
raw_loan_data['click_time'] = raw_loan_data['click_time'].apply(lambda x: x.split('-')[1])
raw_loan_data = raw_loan_data[(raw_loan_data.click_time!= '12')&(raw_loan_data.click_time!= '11')]
new_loan_data = raw_loan_data.groupby('uid', sort=False)\
	.agg({'click_time': len,})\
	.rename(columns={'click_time': 'click_count_sum'})
new_loan_data.to_csv('./featureData/new_click_data.csv')

user_data = pd.read_csv('D:\研究生\Chaos\JD\信贷需求/t_user.csv')
loan_data = pd.read_csv('./featureData/new_click_data.csv')
merge_loan_data = pd.merge(user_data, loan_data, how='left', on='uid')
merge_loan_data.to_csv('./featureData/merge_click_data.csv')
'''t_order'''
raw_loan_data = pd.read_csv('D:\研究生\Chaos\JD\信贷需求/t_order.csv')
raw_loan_data['buy_time'] = raw_loan_data['buy_time'].apply(lambda x: x.split('-')[1])
raw_loan_data['price'] = raw_loan_data['price'].apply(original_data)
raw_loan_data['discount'] = raw_loan_data['discount'].apply(original_data)
raw_loan_data = raw_loan_data[(raw_loan_data.buy_time!= '12')&(raw_loan_data.buy_time!= '11')]
new_loan_data = raw_loan_data.groupby('uid', sort=False)\
	.agg({'buy_time': len, 'price':lambda x: sum(x), })\
	.rename(columns={'buy_time': 'order_count_sum', 'price': 'order_price_sum'})
new_loan_data.to_csv('./featureData/new_order_data.csv')

user_data = pd.read_csv('D:\研究生\Chaos\JD\信贷需求/t_user.csv')
loan_data = pd.read_csv('./featureData/new_order_data.csv')
merge_loan_data = pd.merge(user_data, loan_data, how='left', on='uid')
merge_loan_data.to_csv('./featureData/merge_order_data.csv')

'''user ability'''
user_data = pd.read_csv('D:\研究生\Chaos\JD\信贷需求/t_user.csv')
loan_data = pd.read_csv('./featureData/new_loan_data.csv')
order_data = pd.read_csv('./featureData/new_order_data.csv')
click_data = pd.read_csv('./featureData/new_click_data.csv')
user_data = user_data[['uid']].drop_duplicates(['uid']).reset_index()
user_ability = pd.merge(user_data, order_data, how='left', on='uid')
user_ability = pd.merge(user_ability , click_data, how='left', on='uid')
user_ability = pd.merge(user_ability , loan_data, how='left', on='uid')
user_ability =user_ability.fillna(0)


user_ability['order_ability']= pd.Series(user_ability['order_price_sum'])/pd.Series(user_ability['order_count_sum'])
user_ability['loan_ability'] = pd.Series(user_ability['loan_price_sum'])/pd.Series(user_ability['loan_count_sum'])
user_ability['repayment_ability'] = pd.Series(user_ability['loan_price_sum'])/pd.Series(user_ability['plannum_sum'])
user_ability['active_ability'] = pd.Series(user_ability['click_count_sum'])/3.0
user_ability['consumer_model'] = pd.Series(user_ability['loan_price_sum'])/pd.Series(user_ability['order_price_sum'])
user_ability['click_model'] = pd.Series(user_ability['loan_count_sum'])/pd.Series(user_ability['order_count_sum'])
user_ability =user_ability.replace([np.inf,-np.inf],99999)
user_ability.to_csv('D:/研究生/Chaos/JD/信贷需求/feature11/ability11.csv',index= None)
```
